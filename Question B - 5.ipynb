## Import data

```
df_stock = pd.read_csv('data/FB_data.csv', parse_dates=True, index_col = "Date")

df_stock['Date'] = df_stock.index

df = df_stock

df.index = pd.DatetimeIndex(df.index).to_period('D')
```

# Visualization

```
dates = pd.date_range(start='2010-01-01', end='2022-12-31',periods=253)

fig = px.line(
    data_frame = df_stock, 
    x=dates, 
    y='High', 
    title='Stock price (High) with Rangeslider and RangeSliders'
)

fig.update_xaxes(        
    rangeslider_visible = True,

    rangeselector_buttons = list(
        [          
          dict(count = 3,  label = "1Q", step = "month", stepmode = "backward"),
          dict(count = 6,  label = "2Q", step = "month", stepmode = "backward"),
          dict(count = 9,  label = "3Q", step = "month", stepmode = "backward"), 
          dict(count = 12, label = "4Q", step = "month", stepmode = "backward"), 
          dict(step="all")                                                       
        ] 
    )
)
fig.show()
```

# Visualization for moving average 
with window size 5 days

```
window_size = 5
dates = pd.date_range(start='2010-01-01', end='2022-12-31',periods=253)

df_rolling = df['Open'].rolling(window = window_size).mean()

fig = go.Figure()

fig.add_trace(go.Scatter(x = dates, y = df_rolling, name="moving average"))

fig.update_xaxes(        
    rangeslider_visible = True 
)

fig.update_xaxes(        
    rangeslider_visible = True,

    rangeselector_buttons = list(
        [          
          dict(count = 3,  label = "1Q", step = "month", stepmode = "backward"),
          dict(count = 6,  label = "2Q", step = "month", stepmode = "backward"),
          dict(count = 9,  label = "3Q", step = "month", stepmode = "backward"), 
          dict(count = 12, label = "4Q", step = "month", stepmode = "backward"), 
          dict(step="all")                                                       
        ] 
    )
)

fig.show()
```
